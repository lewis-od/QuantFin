```python
%matplotlib inline
import numpy as np
import quantfin as qf
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(0.0, 1.0, size=(200,))
```

```python
t = np.linspace(-3, 3, 200)
gauss = qf.data.smoothing.kde(t, x, 0.65)
epan = qf.data.smoothing.kde(t, x, 0.8, k=qf.data.dist.epanechnikov)

plt.hist(x, density=True);
gauss_plot, = plt.plot(t, gauss);
epan_plot, = plt.plot(t, epan);
plt.legend([gauss_plot, epan_plot], ["Gaussian", "Epanechnikov"]);
```

```python
P = qf.random_walk.geometric_brownian(250, 0.01, 0.0, 0.4)
R = qf.utils.returns_from_prices(P)
t = np.arange(len(R))
```

```python
simple_avg = qf.data.smoothing.sma(R, 10)
weighted_avg = qf.data.smoothing.ewma(R, 0.2)

plt.plot(t, R);
sma_plot, = plt.plot(t[10:], simple_avg)
ewma_plot, = plt.plot(t, weighted_avg)
plt.legend([sma_plot, ewma_plot], ["SMA", "EWMA"]);
```
